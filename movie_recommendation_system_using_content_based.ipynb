```python
#import or define neccessary library for work on the data

import numpy as np
import pandas as pd
```

```python
#preprocessing of movies data set

movies=pd.read_csv('/content/movie_recommendattion file/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/movie_recommendattion file/tmdb_5000_credits.csv')

movies.head(0) #finding number of columns in movies dataset
```

```python
credits.head(0) #finding number of columns in credits data set
```

```python
movies=movies.merge(credits,on='title')
```

```python
#main columms that we used for movies recommendationn

#genres
#id
#title
#keywords
#overview
#cast
#crew
```

```python
movies=movies[['genres','id','title','keywords','overview','cast','crew']] #extraction of our main and useful columns in the dataset
```

```python
movies.isnull().sum() #find if any of the  column contain null data set
```

```python
movies.dropna(inplace=True) #delete all the those missing values from the dataset .
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#required_output={'Action','Adventure','Fantasy','Science fiction'}
```

```python
#use of ast
#ast.literal_eval is library function which is used to convert the dictionary into a list or a  string
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
#till now we fetch the genres, keyword,cast , crew or can say director from the data and we preprocess the raw data into our meaning full data
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x] )
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x] )
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x] )
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x] )
```

```python
movies.head()
```

```python
#preprocessing is done uptill now and now we are starting with creating tags column from here
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
movies.head(1)
```

```python
new_df=movies[['title','id','tags']]
new_df.head()

#here we get the list form of tags now we have to convert list to string
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) #code for convertinng the list into string
```

```python
new_df.head()
```

```python
pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#from here we remove the stcoks words and convert it into vector
#for this purpose we import from sklearn library
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")#here we use max feature =5000 to select top 5000 words or frequent words from vector and stop_words=english
#for removing stop words in the english alphabet
```

```python
#conversion of tags string to vector
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names() #top 5000 words we get from tags columns
```

```python
from sklearn.metrics.pairwise import cosine_similarity #library for cosine similarity
```

```python
similarity=cosine_similarity(vectors) #cosine similarity between the movies
```

```python
cosine_similarity(vectors).shape
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] #to get the movie index
    distances=similarity[movie_index] #to get the distance or vector on the basis of title
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]#fetching the top 5 movies from the distance vector
    for i in movie_list:
        print(new_df.iloc[i[0]].title) #new_df.iloc[i[0]] display all the information of the movies .tile give only the title of the movies


```

```python
recommend('Avatar')
```
